```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create initial line
line_length = 1
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta) * line_length
y = np.sin(theta) * line_length

fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Define the rotation function
def rotate_line(angle):
    theta_rotated = theta + angle
    x_rotated = np.cos(theta_rotated) * line_length
    y_rotated = np.sin(theta_rotated) * line_length
    line.set_xdata(x_rotated)
    line.set_ydata(y_rotated)
    return line,

# Create animation
ani = animation.FuncAnimation(fig, rotate_line, frames=np.linspace(0, 2*np.pi, 360), interval=50)

plt.axis('equal')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
k = 10
c = 0.5
x_range = np.linspace(-20, 20, 1000)
times = [0, 10, 20]
def func(x, t):
    return np.exp(-(x - c*t)**2) * np.cos(k * (x - c*t))
plt.figure(figsize=(12, 8))
for t in times:
    y_values = func(x_range, t)
    plt.plot(x_range, y_values, label=f't = {t}')
plt.title('Function e^(-(x-ct)^2) cos(k (x - ct)) at Different Times')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
def y(x, n, L):
    k = n * np.pi / L
    return np.sin(k * x)
def integrand(x, n, L):
    return y(x, n, L)**2 * x
L = 3
ns = [1, 2, 3]
plt.figure(figsize=(10, 6))

for n in ns:
    result, _ = quad(integrand, 0, L, args=(n, L))
    constant = 1 / result
    x_values = np.linspace(0, L, 1000)
    y_values = y(x_values, n, L) * np.sqrt(constant)
    plt.plot(x_values, y_values, label=f'n = {n}')

plt.title('Solution of the Differential Equation')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

L = 1.0
x = np.linspace(0, L, 1000)

def wavefunction(n, x):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)
def probability_density(n, x):
    psi = wavefunction(n, x)
    return psi ** 2

n_values = np.arange(0, 51, 5)

for n in n_values:
    plt.figure(figsize=(8, 4))

    plt.subplot(121)
    plt.plot(x, wavefunction(n, x), label=f'Wavefunction (n={n})')
    plt.xlabel('Position (x)')
    plt.ylabel('Wavefunction')
    plt.legend()

    plt.subplot(122)
    plt.plot(x, probability_density(n, x), label=f'Probability Density (n={n})', color='red')
    plt.xlabel('Position (x)')
    plt.ylabel('Probability Density')
    plt.legend()

    plt.suptitle(f'1D Box - n={n}')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def rotate_point(point, angle_degrees, axis='y'):
    angle_radians = np.radians(angle_degrees)
    if axis == 'x':
        rotation_matrix = np.array([[1, 0, 0],
                                    [0, np.cos(angle_radians), -np.sin(angle_radians)],
                                    [0, np.sin(angle_radians), np.cos(angle_radians)]])
    elif axis == 'y':
        rotation_matrix = np.array([[np.cos(angle_radians), 0, np.sin(angle_radians)],
                                    [0, 1, 0],
                                    [-np.sin(angle_radians), 0, np.cos(angle_radians)]])
    elif axis == 'z':
        rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians), 0],
                                    [np.sin(angle_radians), np.cos(angle_radians), 0],
                                    [0, 0, 1]])
    else:
        raise ValueError("Invalid axis. Use 'x', 'y', or 'z'")

    rotated_point = np.dot(rotation_matrix, point)
    return rotated_point

line_start = np.array([0, 0, 0])
line_end = np.array([1, 1, 1])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D([line_start[0], line_end[0]], [line_start[1], line_end[1]], [line_start[2], line_end[2]], label='Original Line')

for angle in range(0, 360, 10):
    rotated_end = rotate_point(line_end, angle, axis='y')
    ax.plot3D([line_start[0], rotated_end[0]], [line_start[1], rotated_end[1]], [line_start[2], rotated_end[2]])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```
